# Highlighting the Maximum Value of each Column

```python
import pandas as pd
```

```python
nba = pd.read_csv('../data/nba_stats.csv')
nba.head(10)
```

### take max of each column

```python
nba.max()
```

### Use boolean selection

```python
nba[nba['STL'] == nba['STL'].max()]
```

### Use idxmax - set_index first and only numeric columns

```python
nba_numeric = nba.set_index('Player').select_dtypes('number')
nba_numeric.head()
```

```python
idx = nba_numeric.idxmax()
idx
```

```python
nba_numeric.loc[idx]
```

```python
nba_numeric = nba.set_index(['Player', 'Year']).select_dtypes('number')
nba_numeric.head()
```

```python
idx = nba_numeric.idxmax().drop_duplicates()
idx
```

```python
nba_idx_max = nba_numeric.loc[idx]
nba_idx_max
```

### Use style attribute

```python
nba_idx_max.style.highlight_max()
```
